```python
df_channels = pd.read_csv('channel_spend_graduate.csv')
df_channels
```

```python
df_channels['channel_date']= df_channels['channel']+str(df_channels['date'])
df_channels
```

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x(), 1.03*height, '%.2f' % float(height))
```

```python
plt.figure(figsize=(25,10))

channel_l = list(df_channels['channel'].unique())

date1907=list(df_channels['spend_AED'][df_channels['date']== 201907])
date1908 = list(df_channels['spend_AED'][df_channels['date']== 201908])
date1909=list(df_channels['spend_AED'][df_channels['date']== 201909])
date1910=list(df_channels['spend_AED'][df_channels['date']== 201910])
date1911=list(df_channels['spend_AED'][df_channels['date']== 201911])
date1912=list(df_channels['spend_AED'][df_channels['date']== 201912])
date2001=list(df_channels['spend_AED'][df_channels['date']== 202001])
date2002=list(df_channels['spend_AED'][df_channels['date']== 202002])

bar_width=0.12

x =np.arange(5)

a=plt.bar(x,date1907,bar_width,label='1907',alpha=0.5)
b=plt.bar(x+bar_width,date1908,bar_width,label='1908',alpha=0.5)
c=plt.bar(x+bar_width+bar_width,date1909,bar_width,label='1909',alpha=0.5)
d=plt.bar(x+bar_width+bar_width+bar_width,date1910,bar_width,label='1910',alpha=0.5)
e=plt.bar(x+bar_width+bar_width+bar_width+bar_width,date1911,bar_width,label='1911',alpha=0.5)
f=plt.bar(x+bar_width+bar_width+bar_width+bar_width+bar_width,date1912,bar_width,label='1912',alpha=0.5)
g=plt.bar(x+bar_width+bar_width+bar_width+bar_width+bar_width+bar_width,date2001,bar_width,label='2001',alpha=0.5)
h=plt.bar(x+bar_width+bar_width+bar_width+bar_width+bar_width+bar_width+bar_width,date2002,bar_width,label='2002',alpha=0.5)

autolabel(a)
autolabel(b)
autolabel(c)
autolabel(d)
autolabel(e)
autolabel(f)
autolabel(g)
autolabel(h)

plt.xlabel("Channel")
plt.ylabel("Spend_aed")

plt.xticks(x+bar_width/1.3,channel_l)

plt.legend()

plt.show()
```
